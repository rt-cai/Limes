```python
import requests as R
import json

URL = "https://elab.msl.ubc.ca/api/v1"
```

```python
TOKEN = ''
USER = {}

def auth():
    CRED = 'credentials/elab.msl' 
    with open(CRED) as cred:
        u, p, _ = [l[:-1] for l in cred.readlines()]
    res = R.post(f"{URL}/auth/user/", json={
        "username": u,
        "password": p
    })

    try:
        global TOKEN, USER
        data = json.loads(res.text)
        TOKEN = data['token']
        USER = data['user']
        return USER['firstName'], USER['lastName']
    except:
        return res.text
print(auth())
```

```python
def mh():
    return {
        'Authorization': TOKEN
    }

def call(url, method=R.get, jn={}):
    res = method(url, headers=mh(), json=jn)
    try:
        data = json.loads(res.text)
        return data
    except:
        return res.status_code, res.text
```

```python

name="MMAP-Sample"

x = call(f"{URL}/sampleTypes?name={name}", R.get)
x
```
